```python
#missing value handling
for i in ('LoanAmount',
       'Loan_Amount_Term', 'Credit_History'):
  data[i].fillna(data[i].median(),inplace=True)
```

```python
for i in('Gender','Dependents','Credit_History','Self_Employed' ):
  data[i].fillna(data[i].mode()[0],inplace=True)
```

```python
#one hot encording
cols=[ 'Gender', 'Married', 'Dependents', 'Education',
       'Self_Employed','Property_Area']
data=pd.get_dummies(data,columns=cols,dtype='int64')
df=pd.get_dummies(df,columns=cols,dtype='int64')
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df['Loan_Status']=le.fit_transform(df['Loan_Status'])
```

```python
x=df.drop('Loan_Status',axis=1)
y=df['Loan_Status']
```

```python
x=x.drop('Loan_ID',axis=1)
```

```python
x.head()
```

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression(max_iter=1000)
```

```python
lr.fit(x,y)
```

```python
y_pred=lr.predict(data)
```

```python
y_pred
```

```python
file=pd.DataFrame(y_pred,columns=['Loan_Status'])
```

```python
file=file.replace({1:'Y',0:'N'})
```

```python
submission=pd.concat([id,file],axis=1)
submission.to_csv('submission.csv',index=False)
```
