```python
#scaling the data
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
```

```python
columns_scale=['Fresh','Milk']
data[columns_scale]=sc.fit_transform(data[columns_scale])
```

```python
data.head()
```

```python
#creating model
x=data[['Fresh','Milk']]
x.head()
```

```python
from sklearn.cluster import KMeans
```

```python
#specifing the cluster with wcss
#k-mean cluster
wcss=[]
for i in range(1,11):
  km=KMeans(n_clusters=i,n_init=10)
  km.fit(x)
  css=km.inertia_
  wcss.append(css)
```

```python
plt.plot(range(1,11),wcss)
plt.xlabel('number of cluster')
plt.ylabel('wcss')
plt.grid()
```

```python
km=KMeans(n_clusters=7,n_init=10)
km.fit(x)
```

```python
cluster_labele=km.predict(x)
```

```python
pd.Series(cluster_labele)
data['cluster']=cluster_labele
```

```python
plt.scatter(x['Fresh'], x['Milk'], c=cluster_labele)
plt.xlabel('Fresh')
plt.ylabel('Milk')
plt.grid()
```

```python
#prediction
def predict_cluster(Fresh,Milk):
  new_data=[[Fresh,Milk]]
  cluster=km.predict(new_data)
  return cluster[0]
```

```python
# new data
new_rhob=-0.30
new_nphi=0.40
c = predict_cluster(new_rhob, new_nphi)
print(f'The new data belongs to cluster{c}')
```

```python
#agglomerative hc
from scipy.cluster.hierarchy import linkage
from scipy.cluster.hierarchy import dendrogram
from scipy.cluster.hierarchy import fcluster
```

```python
mergings=linkage(x,method='ward')
```

```python
plt.figure(figsize=(15,10))
dendrogram(mergings)
plt.xlabel('sample')
plt.ylabel('distance')
plt.show()
```

```python
cluster_labels = fcluster(mergings, t=150, criterion='distance')
```
